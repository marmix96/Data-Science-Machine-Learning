<li>Να ταξινομήσετε τις καταστάσεις της αλυσίδας σε κλάσεις επικοινωνίας. Ποιες κλάσεις είναι ανοιχτές και ποιες κλειστές; Ποιες καταστάσεις είναι παροδικές και ποιες επαναληπτικές; Για να βοηθηθείτε μπορείτε να χρησιμοποιήσετε τον <a href="https://setosa.io/markov/index.html#%7B%22tm%22%3A%5B%5B0.5%2C0.5%2C0%2C0%2C0%2C0%2C0%2C0%5D%2C%5B0.25%2C0.75%2C0%2C0%2C0%2C0%2C0%2C0%5D%2C%5B0.25%2C0.25%2C0%2C0.125%2C0.375%2C0%2C0%2C0%5D%2C%5B0%2C0%2C0.25%2C0%2C0.75%2C0%2C0%2C0%5D%2C%5B0%2C0%2C0.2%2C0.2%2C0.2%2C0.2%2C0.2%2C0%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0%2C0.5%2C0.5%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0.5%2C0%2C0.5%5D%2C%5B0%2C0%2C0%2C0%2C0%2C0.5%2C0.5%2C0%5D%5D%7D">προσομοιωτή</a>, ο οποίος είναι κατάλληλα διαμορφωμένος για την άσκηση.</li>

Παρατηρούμε ότι ο υποπίνακας των καταστάσεων 1 και 2 είναι στοχαστικός, άρα αυτές σχηματίζουν την πρώτη κλειστή κλάση, $C_1 = \{1, 2\}$. Το ίδιο παρατηρούμε για τις καταστάσεις 6, 7 και 8 οι οποίες δημιουργούν επίσης κλειστή κλάση, έστω $C_3 = \{6, 7, 8\}$. Οι υπολειπόμενες καταστάσεις 3, 4 και 5 αφού επικοινωνούν μεταξύ τους ορίζουν μια κλάση, έστω $C_2 = \{3, 4, 5\}$. Η αλυσίδα όμως μπορεί να φύγει από την $C_2$ επομένως πρόκειται για ανοιχτή κλάση. Οι καταστάσεις κλειστών κλάσεων είναι επαναληπτικές, δηλαδή οι καταστάσεις $C_1 \cup C_3 = \left\{1,2,6,7,8\right\}$, ενώ οι καταστάσεις ανοιχτών κλάσεων είναι παροδικές, δηλαδή στην συγκεκριμένη περίπτωση οι καταστάσεις $C_2 = \{3, 4, 5\}$.

<li>Χρησιμοποιώντας το αποτέλεσμα του προηγούμενου ερωτήματος, να προσδιορίσετε την αναλλοίωτη κατανομή της μαρκοβιανής αλυσίδας για την περίπτωση που η αλυσίδα ξεκινάει από την κατάσταση (α) 1 και (β) 8, χωρίς τη βοήθεια υπολογιστή.</li>

(α) Λύνουμε το πρόβλημα στον υποπίνακα της κλειστής κλάσης $C_1 = \{1, 2\}$. Προφανώς, $\pi_3 = \cdots = \pi_8 = 0$. Για την αναλοίωτη κατανομή θα ισχύει:
$$
\begin{equation}
        \begin{aligned}
            \pi = \pi P \Rightarrow \pi_{1} = 0.5\pi_{1} + 0.25\pi_{2} \text{, } \pi_{2} = 0.5\pi_{1} + 0.75\pi_{2} \textbf{ και } \pi_{1}+\pi_{2} = 1
        \end{aligned}
    \end{equation}
\implies
\pi_1  = \frac{1}{3}, \text{ }\pi_2  = \frac{2}{3}
$$
Επομένως λαμβάνουμε την εξής αναλλοίωτη κατανομή:
$$ \pi = \left(\frac{1}{3}, \frac{2}{3}, 0, 0, 0, 0, 0, 0\right)$$

(β) Από την παρατήρηση ότι ο υποπίνκας της $C_3 = \{6, 7, 8\}$ είναι διπλά στοχαστικός έπεται αμέσως $$ \pi = \left(0, 0, 0, 0, 0, \frac{1}{3}, \frac{1}{3}, \frac{1}{3}\right)$$.

<li> Να θεωρήσετε την περίπτωση στην οποία η αλυσίδα ξεκινάει από την κατάσταση 1. Να υποθέσετε ότι σε κάποιο βήμα, η αλυσίδα βρίσκεται στην κατάσταση 2 και ότι έχει παρέλθει το μεταβατικό φαινόμενο. Ποιος είναι ο αναμενόμενος αριθμός βημάτων που απαιτούνται για την πρώτη επιστροφή της αλυσίδας
στην κατάσταση 2; Να απαντήσετε στο ερώτημα χωρίς τη βοήθεια υπολογιστή.</li>

Δεδομένου ότι η αρχική κατάσταση της αλυσίδας είναι η κατάσταση 1, όπως επεξηγήθηκε αναλυτικά στο προηγούμενο ερώτημα, θα εγκλωβιστούμε εντός της κλειστής κλάσης επικοινωνίας $C_1$. Σε αυτή την περίπτωση γνωρίζουμε ότι ο αναμενόμενος αριθμός βημάτων επανεπιστροφής στην κατάσταση 2, θα είναι:
\begin{equation}
    \begin{aligned}
        \mathbb{E}(τ_{22}) = \frac{1}{\pi_{2}} = \frac{3}{2} \text{ βήματα}
    \end{aligned}
\end{equation}
όπου $\pi_{i} \text{ } \forall i \in \mathbb{X}$, η αναλλοίωτη κατανομή που προέκυψε από το προηγούμενο ερώτημα.
