#### Ακολούθως, γίνονται τα απαραίτητα imports των βιβλιοθηκών

#### Φόρτωση δεδομένων και απομόνωση στηλών που μας ενδιαφέρουν

```python
data1a = pd.read_csv('data1a.csv')
data1a = data1a[["likes", "views"]]
# print(data1a)
```

#### Σε αυτό το σημείο υπολογίζουμε την συσχέτιση μεταξύ εξαρτημένης και ανεξάρτητης μεταβλητής για τα δύο datasets, προκειμένου να εξακριβωθεί ο βαθμός γραμμικής εξάρτησης ανάμεσα στις δύο μεταβλητές

```python
# calculate & print the correlation for the 2 cases

print("Correlation between likes and views (data1a.csv) is: {:.5f}".format(data1a["likes"].corr(data1a["views"]))) # 1st dataset
print("Correlation between x and y (data1b.csv) is: {:.5f}".format(data1b["x"].corr(data1b["y"])))  # 2nd dataset
```

Αρχικά, παρατηρούμε ότι η τιμή της συσχέτισης και στις 2 περιπτώσεις είναι θετική και σχετικά κοντά στην μονάδα. Το πρώτο σημαίνει ότι η εξαρτημένη και η ανεξάρτηση μεταβλητή μεταβάλλονται παρομοίως. Στην πρώτη περίπτωση, η γραμμική εξάρτηση δεν είναι τόσο ισχυρή, ωστόσο είναι μεγαλύτερη του 80%, οπότε θα μπορούσαμε να πούμε ότι ειναι σχετικά ικανοποιητική. ### Διαχωρισμός δεδομένων σε train set και test set

```python
# split randomly the data for the two cases (80% Train Set & 20% Test Set)
X_train_1a, X_test_1a, y_train_1a, y_test_1a = train_test_split(data1a["likes"].values, data1a["views"].values, test_size = 0.2)
X_train_1b, X_test_1b, y_train_1b, y_test_1b = train_test_split(data1b["x"].values, data1b["y"].values, test_size = 0.2)
```

```python
X_train_1a = X_train_1a.reshape(-1, 1)
X_test_1a = X_test_1a.reshape(-1, 1)

X_train_1b = X_train_1b.reshape(-1, 1)
X_test_1b = X_test_1b.reshape(-1, 1)
```

### Κανονικοποίηση των δεδομένων

```python
scaler_a = preprocessing.MinMaxScaler()
scaler_a.fit(X_train_1a)
X_train_1a_scaled = scaler_a.transform(X_train_1a)
X_test_1a_scaled = scaler_a.transform(X_test_1a)

y_train_1a_scaled = scaler_a.transform(y_train_1a.reshape(-1, 1)).reshape(-1)
y_test_1a_scaled = scaler_a.transform(y_test_1a.reshape(-1, 1)).reshape(-1)
```

```python
scaler_b = preprocessing.MinMaxScaler()
scaler_b.fit(X_train_1b)
X_train_1b_scaled = scaler_b.transform(X_train_1b)
X_test_1b_scaled = scaler_b.transform(X_test_1b)

y_train_1b_scaled = scaler_b.transform(y_train_1b.reshape(-1, 1)).reshape(-1)
y_test_1b_scaled = scaler_b.transform(y_test_1b.reshape(-1, 1)).reshape(-1)
```

### SGD_algorithm

```python
def SGD_train(X_train, y_train, X_test, y_test, learning_rate = 0.01, epochs = 100):
    
    # Initially we will keep our W and B as 0 as per the Training Data
    w = np.zeros(shape = (1, X_train.shape[1]))
    b = 0
    mse = np.zeros(epochs)
    
    # for epoch in tqdm(range(epochs), total = epochs):
    for epoch in range(epochs):
        # We keep our initial gradients as 0
        w_gradient = np.zeros(shape = (1, X_train.shape[1]))
        b_gradient = 0
        
        prediction = np.zeros(len(X_train))
        for i in range(len(X_train)):
            prediction[i] = np.dot(w, X_train[i]) + b
            w_gradient = w_gradient + (-2) * X_train[i] * (y_train[i] - prediction[i])
            b_gradient = b_gradient + (-2) * (y_train[i] - prediction[i])
        
        #Updating the weights(W) and Bias(b) with the above calculated Gradients
        w = w - learning_rate * (w_gradient / (len(X_train)))
        b = b - learning_rate * (b_gradient / (len(X_train)))
        
        y_pred = w[0, 0] * X_test + b
        mse[epoch] = mean_squared_error(y_test, y_pred)
        
        # mse[epoch] = mean_squared_error(
            # scaler_a.inverse_transform(y_test.reshape(-1, 1)).reshape(-1),
            # scaler_a.inverse_transform(y_pred.reshape(-1, 1)).reshape(-1)
        # )
        
    return w, b, mse # Returning the weights and Bias and the mean squared error per epoch

def SGD_predict(X_test, w, b):
    y_pred= []
    for i in range(len(X_test)):
        y = np.dot(w, X_test[i]) + b
        y_pred.append(y)
    return np.array(y_pred)
  
```

```python
fig, axs = plt.subplots(1, 2, figsize = (20, 10))
axs = axs.ravel()
data = [
    (X_train_1a_scaled, y_train_1a_scaled, X_test_1a_scaled, y_test_1a_scaled),
    (X_train_1b_scaled, y_train_1b_scaled, X_test_1b_scaled, y_test_1b_scaled)
]
for ax, d, l in zip(axs, data, ["a", "b"]):
    w, b, mse = SGD_train(*d, epochs = 10000)
    y_pred = SGD_predict(d[2], w, b)
    print("Mean squared error = {}".format(mean_squared_error(y_pred, d[3])))
    ax.plot(range(1, len(mse) + 1), mse)
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Mean Squared Error")
    ax.set_title("MSE over epochs for dataset 1{}".format(l))
plt.tight_layout()
```

Παρατηρούμε ότι στο πρώτο dataset δεν ήταν αρκετές ούτε οι 10000 εποχές, που είναι ένας αρκετά μεγάλος αριθμός επαναλήψεων. Αντίθετα, στο δεύτερο έχουμε εξαιρετικά αποτελέσματα. Συνεπώς, επιβεβαιώνουμε τις παρατηρήσεις μας με βάση το correlation των μεταβλητών. Στην πρώτη περίπτωση, εντοπίζουμε ότι υπάρχει κάποια γραμμική εξάρτηση που προσπαθεί να εντοπίσει ο αλγόριθμος, ωστόσο δεν είναι αρκετά ισχυρή ώστε να μπορέσει να συγκλίνει ακόμη και μετά από 10000 εποχές. Ενώ, στην δέυτερη περίπτωση η σύγκλιση έχει επιτευχθεί ήδη απο τις 1800 περίπου εποχές. Τα πειράματα έγιναν με σταθερό ρυθμό μάθησης ίσο με 0.01. Κατόπιν, θα αξιολογήσουμε την επίδραση του learning rate για διαφορετικές τιμές, μελετώντας την επίδραση της συγκεκριμένης παραμέτρου. Για τον σκοπό αυτό θα χρησιμοποιήσουμε το δεύτερο dataset, διότι το πρώτο πιθανώς να μην μας παρέχει αξιόπιστα αποτελέσματα.

```python
for lr in [0.001, 0.005, 0.01, 0.05, 0.1, 0.5]:
    w, b, mse = SGD_train(*data[1], learning_rate = lr, epochs = 500)
    y_pred = SGD_predict(data[1][2], w, b)
    print("Mean squared error for learning rate = {}: {}".format(lr, mean_squared_error(y_pred, data[1][3])))
    plt.plot(range(1, len(mse) + 1), mse, label = "lr = {}".format(lr))
plt.title("Learning rates")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.legend(loc = "best", bbox_to_anchor = (1, 0.5))
plt.show()
```

Δεδομένου ότι το 2ο dataset είναι σαφώς ορισμένη η γραμμική εξάρτηση, παρατηρούμε ότι με την αύξηση του learning rate ο αλγόριθμος συγκλίνει ταχύτερα. Για learning rate = 0.05 και πάνω παρατήρουμε ότι η σύγκλιση επιτυγχάνεται περίπου με τον ίδιο ρυθμό και στο ίδιο error. Ακολούθως, θα διεξάγουμε το ίδιο πείραμα, αυτή τη φορά σε μη κανονικοποιημένα δεδομένα.

```python
w, b , mse = SGD_train(X_train_1b, y_train_1b, X_test_1b, y_test_1b, epochs = 20)
y_pred = SGD_predict(X_test_1b, w, b)
print("Mean squared error for non-normalized data = {}".format(mean_squared_error(y_pred, y_test_1b)))
plt.plot(range(1, len(mse) + 1), mse)
plt.title("Training without scaling")
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.show()
```

Το mean squared error σε αυτή την περίπτωση είναι της τάξης του $10^{76}$. Χωρίς scaling λοιπόν, εμφανίζεται το φαινόμενο της έκρηξης των gradinets, το οποίο είναι εμφανές ακόμη και στις 20 εποχές.
