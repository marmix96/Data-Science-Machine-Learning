<h1><b><i>Ο αλγόριθμος Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>)</b></h1>

Στην άσκηση αυτή θα μελετήσετε τον αλγόριθμο της <i>Προσομοιωμένης Ανόπτησης</i> (<i>Simulated Annealing</i>), περισσότερες πληροφορίες για τον οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Simulated_annealing">εδώ</a>.Θα χρησιμοποιήσετε τα <b>δύο έτοιμα</b> προγράμματα που δίνονται στη συνέχεια. Τα προγράμματα αυτά επιλύουν το πρόβλημα του <i>Περιπλανώμενου Πωλητή</i> (<i>Travelling Salesman Problem</i>, <i>TSP</i>), περισσότερες πληροφορίες για το οποίο μπορείτε να βρείτε <a href="https://en.wikipedia.org/wiki/Travelling_salesman_problem">εδώ</a>. Το πρώτο πρόγραμμα επιλύει το πρόβλημα <i>TSP</i> χρησιμοποιώντας τον αλγόριθμο <i>Simulated Annealing</i> και το δεύτερο επιλύει το ίδιο πρόβλημα, χρησιμοποιώντας τον αλγόριθμο <a href="https://en.wikipedia.org/wiki/Hill_climbing"><i>Hill Climbing</i></a>. Η επίλυση του προβλήματος γίνεται για <b>150 πόλεις</b>.

<h2><b><i>Simulated Annealing</i></b></h2>

```python
#!/usr/bin/env python

""" Traveling salesman problem solved using Simulated Annealing.
http://www.physics.rutgers.edu/~haule/681/src_MC/python_codes/salesman.py
"""
from scipy import *
from pylab import *
from time  import time

def Distance(R1, R2):
    return sqrt((R1[0]-R2[0])**2+(R1[1]-R2[1])**2)

def TotalDistance(city, R):
    dist=0
    for i in range(len(city)-1):
        dist += Distance(R[city[i]],R[city[i+1]])
    dist += Distance(R[city[-1]],R[city[0]])
    return dist
    
def reverse(city, n):
    nct = len(city)
    nn = (1+ ((n[1]-n[0]) % nct))/2 # half the lenght of the segment to be reversed
    # the segment is reversed in the following way n[0]<->n[1], n[0]+1<->n[1]-1, n[0]+2<->n[1]-2,...
    # Start at the ends of the segment and swap pairs of cities, moving towards the center.
    for j in range(nn):
        k = (n[0]+j) % nct
        l = (n[1]-j) % nct
        (city[k],city[l]) = (city[l],city[k])  # swap
    
def transpt(city, n):
    nct = len(city)
    
    newcity=[]
    # Segment in the range n[0]...n[1]
    for j in range( (n[1]-n[0])%nct + 1):
        newcity.append(city[ (j+n[0])%nct ])
    # is followed by segment n[5]...n[2]
    for j in range( (n[2]-n[5])%nct + 1):
        newcity.append(city[ (j+n[5])%nct ])
    # is followed by segment n[3]...n[4]
    for j in range( (n[4]-n[3])%nct + 1):
        newcity.append(city[ (j+n[3])%nct ])
    return newcity

def Plot(city, R, dist):
    # Plot
    Pt = [R[city[i]] for i in range(len(city))]
    Pt += [R[city[0]]]
    Pt = array(Pt)
    title('Total distance='+str(dist))
    plot(Pt[:,0], Pt[:,1], '-o')
    show();

if __name__=='__main__':

    ncity = 150        # Number of cities to visit
    maxTsteps = 100    # Temperature is lowered not more than maxTsteps
    Tstart = 0.2       # Starting temperature - has to be high enough
    fCool = 0.9        # Factor to multiply temperature at each cooling step
    maxSteps = 100*ncity     # Number of steps at constant temperature
    maxAccepted = 10*ncity   # Number of accepted steps at constant temperature

    Preverse = 0.5      # How often to choose reverse/transpose trial move

    # Choosing city coordinates
    R=[]  # coordinates of cities are choosen randomly
    for i in range(ncity):
        R.append( [rand(),rand()] )
    R = array(R)

    # The index table -- the order the cities are visited.
    city = range(ncity)
    # Distance of the travel at the beginning
    dist = TotalDistance(city, R)

    # Stores points of a move
    n = zeros(6, dtype=int)
    nct = len(R) # number of cities
    
    T = Tstart # temperature

    Plot(city, R, dist)
    
    start = time()
    for t in range(maxTsteps):  # Over temperature

        accepted = 0
        for i in range(maxSteps): # At each temperature, many Monte Carlo steps
            
            while True: # Will find two random cities sufficiently close by
                # Two cities n[0] and n[1] are choosen at random
                n[0] = int((nct)*rand())     # select one city
                n[1] = int((nct-1)*rand())   # select another city, but not the same
                if (n[1] >= n[0]): n[1] += 1   #
                if (n[1] < n[0]): (n[0],n[1]) = (n[1],n[0]) # swap, because it must be: n[0]<n[1]
                nn = (n[0]+nct -n[1]-1) % nct  # number of cities not on the segment n[0]..n[1]
                if nn>=3: break
        
            # We want to have one index before and one after the two cities
            # The order hence is [n2,n0,n1,n3]
            n[2] = (n[0]-1) % nct  # index before n0  -- see figure in the lecture notes
            n[3] = (n[1]+1) % nct  # index after n2   -- see figure in the lecture notes
            
            if Preverse > rand(): 
                # Here we reverse a segment
                # What would be the cost to reverse the path between city[n[0]]-city[n[1]]?
                de = Distance(R[city[n[2]]],R[city[n[1]]]) + Distance(R[city[n[3]]],R[city[n[0]]]) - Distance(R[city[n[2]]],R[city[n[0]]]) - Distance(R[city[n[3]]],R[city[n[1]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    reverse(city, n)
            else:
                # Here we transpose a segment
                nc = (n[1]+1+ int(rand()*(nn-1)))%nct  # Another point outside n[0],n[1] segment. See picture in lecture nodes!
                n[4] = nc
                n[5] = (nc+1) % nct
        
                # Cost to transpose a segment
                de = -Distance(R[city[n[1]]],R[city[n[3]]]) - Distance(R[city[n[0]]],R[city[n[2]]]) - Distance(R[city[n[4]]],R[city[n[5]]])
                de += Distance(R[city[n[0]]],R[city[n[4]]]) + Distance(R[city[n[1]]],R[city[n[5]]]) + Distance(R[city[n[2]]],R[city[n[3]]])
                
                if de<0 or exp(-de/T)>rand(): # Metropolis
                    accepted += 1
                    dist += de
                    city = transpt(city, n)
                    
            if accepted > maxAccepted: break
        end = time()
        # Plot
        Plot(city, R, dist)
            
        print "T=%10.5f , distance= %10.5f , accepted steps= %d" %(T, dist, accepted)
        print "Execution time: %f"%(end - start)
        T *= fCool             # The system is cooled down
        if accepted == 0: break  # If the path does not want to change any more, we can stop

        
    Plot(city, R, dist)
```

### *Απαντήσεις*
- Το πρόβλημα του πλανόδιου πωλήτη αποτελεί ένα από τα κλασικότερα προβλήματα συνδιαστικής βελτιστοποίησης και αφορά την εύρεση της βραχύτερης κλειστής διαδρομής, ανάμεσα σε ένα σύνολο  n ομο-επίπεδων σημείων/κόμβων, τα οποία παριστάνουν τις διαφορετικές πόλεις όπου επιθυμεί να επισκευθεί ένας πλανόδιος πωλητής. Ειδικότερα το TSP μπορεί να μοντελοποιηθεί ως μη-κατευθυντικό βεβαρημένο γράφημα, οι κόμβοι του οποίου αναπαριστούν τις πιθανές πόλεις προς επίσκεψη και οι ακμές του τα υφιστάμενα μονοπάτια μεταξύ των πόλεων, όπου το βάρος κάθε ακμής ισούται με την ευκλείδια απόσταση του αντίστοιχου μονοπατιού. Απλούστερα, δεδομένου ενός γράφου με βάρη το πρόβλημα TSP αφορά στην εύρεση μίας περιήγησης που περιέχει όλες τις κορυφές και είναι ελαχίστου βάρους. Γενικά μιλώντας, το πρόβλημα ανήκει στην κατηγορία των NP-complete προβλημάτων, πράγμα που συνεπάγεται ότι η εύρεση της βέλτιστης λύσης του, απαιτεί υπερπολυωνιμικό χρόνο. Παρόλο αυτά, πρόκειται για ένα ευρέως διαδεδομένο πρόβλημα, μιάς και με βάση αυτό, μπορούν να μοντελοποιηθούν αρκετά άλλα, εντελώς διαφορετικά προβλήματα, τα οποία συχνά μπορεί και να προέρχονται από διάφορες επιστημονικές περιοχές. Κατά κύριο λόγο, στην πράξη, χρησιμοποιούνται προσεγγιστικοί αλγόριθμοι για την επίλυση του, οι οποίοι οδηγούν σε κάποιο αρκετά ικανοποιητικό αποτέλεσμα (κατά την πληοψηφία των περιπτώσεων δεν είναι βέλτιστο), με αισθητά χαμηλότερο υπολογιστικό κόστος. Στο πλαίσιο του συγκεκριμένου εργαστηρίου, λοιπόν, επιδιώκουμε την επίλυση του TSP χρησιμοποιώντας δύο κλασσικούς προσεγγιστικούς αλγορίθμους, τον Hill-Climbing και τον Simulated-Anneling αλγόριθμο. 

- Ο αλγόριθμος Hill-Climbing είναι μία επαναληπτική μέθοδος, που χρησιμοποιείται ευρέως για την προσέγγιση του σημείου ελαχίστου μιας δοσμένης συνάρτησης κόστους. Ανήκει στην οικογένεια αλγορίθμων τοπικής αναζήτησης, καθώς ξεκιναέι από μία, αυθαίρετα επιλεγμένη, εφικτή λύση του προβλήματος, προσπαθώντας σε κάθε βήμα να βρεί κάποια καλύτερη λύση εντός της "γειτονιάς" N στην οποία βρίσκεται, σε σχέση με εκείνη την λύση που ήδη διαθέτει. Εάν επιτύχει στην εύρεση καλύτερης λύσης μεταβαίνει εκεί, διαφορετικά η μέθοδος τερματίζει. Ο αλγόριθμος Hill-Climbing βρίσκει το καλύτερο αποτέλεσμα σε περιπτώσεις που η συνάρτηση προς ελαχιστοποίηση είναι αυστηρά κυρτή. Σε γενικές γράμμες, για οποιαδήποτε άλλη περίπτωση, όπως η πλειοψηφιά των ευριστικών αλγορίθμων, εντοπίζει κάποιο τοπικό ελάχιστο, ανάλογα με το σημείο εκκίνησης του. Αυτός είναι και ο λόγος για τον οποίο ο συγκεκριμένος αλγόριθμός μπορεί να οδηγησεί ορισμένες φορές σε κακά αποτελέσματα. Για την εν μέρει αντιμετώπιση αυτού του προβλήματος, απαιτούνται πιο πολύπλοκοι αλγόριθμοι, όπως ο Simulated-Annealing που συζητείται στο ακόλουθο ερώτημα, οι οποίοι προνοούν και ως κάποιο βαθμό αντιμετωπίζουν αντίστοιχες περιπτώσεις.

- Ο αλγόριθμος Simulated Annealing(SA) στηρίζεται στον αλγόριθμο Metropolis-Hastings(MH) που είδαμε στο προηγούμενο εργαστήριο. Για κάποια αρχική θερμοκρασία βρίσκουμε τη στάσιμη κατανομή μέσω του MH και έπειτα ψύχουμε λίγο το σύστημα και ξαναλύνουμε με MH εώς ότου η θερμοκρασία φτάσει στο 0 (ή κάποιο άλλο κριτήριο σύγκλισης πληρωθεί). Το πλεονέκτημα του SA είναι η πιθανοτική φύση του MH. Χειρότερες καταστάσεις μπορεί να επιλέγουν, αν και με μικρότερη πιθανότητα, και να βοηθήσουν να αποφύγουμε ή να ξεκολλήσουμε από ένα τοπικό ελάχιστο. Αντίθετα, ο Hill Climbing πάντα θα επιλέγει την λίγο καλύτερη, αν αυτή υπάρχει, οπότε θα εγκλωβίζεται συχνότερα στα τοπικά ελάχιστα.

- Τα αποτέλεσματα φαίνονται παραπάνω.

- Η καλύτερη λύση επιλέγεται από τον αλγόριθμο Simulated Annealing ωστόσο δεν ειναι πάρα πολύ καλύτερη. Από την άλλη ο Hill Climbing είναι πιο γρήγορος γιατί τερματίζει σε αρκετά υψηλότερη θερμοκρασία. Επίσης, βλέπουμε ότι πιο γρήγορα πλησίαζει στη τελική λύση που επιλέγεται ο Hill Climbing μιας και σε 3-4 βήματα απέχει η απόσταση της τρέχουσας λύσης του(\~11) απέχει λίγοτερο από μονάδα από την τελική του λύση(~10.5). Από την άλλη ο SA πλησιάζει πολύ πιο αργά. Ενδεικτικά, στην θερμοκρασία τερματισμού του Hill Climbing έχει 50% μεγαλύτερη τιμή απόστασης.

- Ο Hill Climbing κάνει μία πολύ έντονη αλλαγή στο πρώτο βήμα και έπειτα κάνει κάποιες πολύ μικρές εώς αμελητέες. Αντίθετα, ο SA κάνει κάποιες έντονες αλλαγές στα πρώτα του βήματα πριν καταλήξει και αυτός σε μικρότερες. Όπως είπαμε και παραπάνω, ο Hill Climbing πάντα επιλέγει καλύτερες ενώ με μικρή πιθανότητα ο SA μπορεί και χειρότερες, π.χ. εδώ πήγε από 15 σε 17.
